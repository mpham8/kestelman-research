Import Libraries

```python
import os
import re
```

OS

```python
folder_path = 'LA-Audios' #SET FOLDER PATH

name_ls = []
def get_file(file_path):
  with open(file_path, 'r') as file:
      text = file.read()
      matches = re.findall(r'my name is (\w+) (\w+)', text, re.IGNORECASE)
      for match in matches:
          name_ls.append(match)


files = os.listdir(folder_path)
```

Call

```python
for file in files:
  get_file(folder_path + "/" + file)

print(f"names: {len(name_ls)}")
name_ls
```
