```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from collections import defaultdict
```

```python
cases_df = pd.read_csv("merge-check-error.csv")
cases_df.head()
```

Is there a specific APC that doesn’t merge?

```python
bad_merges_apc = defaultdict(int)
all_apc = defaultdict(int)

for index, row in cases_df.iterrows():
    if pd.isna(row['proposed_adding']):
        bad_merges_apc[row['area']] += 1
    all_apc[row['area']] += 1

bad_merges_apc
```

```python
all_apc
```

Is there a specific type application reason or entitlement type that does not show up?

```python
bad_merges_type = defaultdict(int)
all_type = defaultdict(int)

for index, row in cases_df.iterrows():
    if pd.isna(row['proposed_adding']):
        bad_merges_type[str(row['cleaned_case_number'])[:3]] += 1
    all_type[str(row['cleaned_case_number'])[:3]] += 1

bad_merges_type
```
